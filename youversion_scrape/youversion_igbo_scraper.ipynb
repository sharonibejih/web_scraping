Following the template url for youversion web bible, which is: https://www.bible.com/bible/77/{book_abbrv}.{chpt}.IGBOB, the previous idea was:
`Extract the first three letters in each book name and convert to upper case.`

But there was inconsistencies. For example, Song of Solomon is SNG and not SON, Ezekiel is EZK and not EZE.

Before opting to create a new list of all youversion abbreviations manually, I luckily stumbled upon this github gist: https://gist.github.com/danott/615135. It contains a JSON of bible names and their youversion abbreviations. 

Huge thanks to Danott and other contributors. I added a ⭐.

```python
old_testament_chpts = [50, 40, 27, 36, 34, 24, 21, 4, 31, 34, 22, 25, 29, 36, 10, 13, 10, 42, 150, 31, 12, 8, 66, 52, 5, 48, 12, 14, 3, 9, 1, 4, 7, 3, 3, 3, 2, 14, 4]
print(len(old_testament_chpts))
```

```python
new_testament_chpts = [28, 16, 24, 21, 28, 16, 16, 13, 6, 6, 4, 4, 5, 3, 6, 4, 3, 1, 13, 5, 5, 3, 5, 1, 1, 1, 22]
print(len(new_testament_chpts))
```

```python
old_testament_abbrev = list(youversion_old_testament_books_abbrev.keys())
new_testament_abbrev = list(youversion_new_testament_books_abbrev.keys())
```

```python
## convert each testament books and their respective chapters to a dictionary
 
old_testament = dict(zip(old_testament_abbrev, old_testament_chpts))
new_testament = dict(zip(new_testament_abbrev, new_testament_chpts))
 
# printing resultant dictionary
print(old_testament)
```

```python
# read text file. all scraped data will be updated to this file
text_file = open("/content/drive/MyDrive/igbo_lm/youversion-new-testament-igbo.txt", "w")
```

```python
## loop through either old or new_testament.items()
for book, chpts in new_testament.items():
  for chpt in range(1, chpts+1):
    book_abbrv = book.upper()

    url = f"https://www.bible.com/bible/77/{book_abbrv}.{chpt}.{version}" # youversion bible url
    print(url)

    html = requests.get(url).text
    body = fulltext(html)
    print(body)
    # append new data to text file
    text_file.write(body+"\n")

text_file.close()
```

A few books like `JOB, MAL, ROM`, however, did not scrape. If you need all the books, you might have to rework the codes. 

## How to reproduce for other languages/versions

1. Choose a version from here:
https://www.bible.com/versions

In this project, I worked with this: https://www.bible.com/versions/77-igbob-bible-nso 

2. When you open the version of your choice, like mine for example, the first two links you will see are: Read Version and Audio Version. Select the Read Version and that's mostly it. It should look similar to this, for Genesis 1: https://www.bible.com/bible/77/GEN.1.IGBOB 

3. The url format used in this project `url = f"https://www.bible.com/bible/77/{book_abbrv}.{chpt}.{version}"` is the same across all versions. Simply ensure that the `version` variable defined at the beginning of the notebook is changed to fit into yours. Mine is `IGBOB`.
